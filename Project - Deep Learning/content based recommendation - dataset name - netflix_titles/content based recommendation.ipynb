```python
import pandas as pd
import numpy as np
from nltk.sem.chat80 import items
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Sample item data (could be movies, products, etc.)
#items = [
 #  "Red running shoes for men",
  #  "Blue running shoes for women",
   # "Formal black leather office shoes",
  #  "Comfortable running sneakers for gym",
  #  "Elegant black leather formal shoes"
#]
items=pd.read_csv('netflix_titles.csv')
items['director'] = items['director'].fillna("")
items['description'] = items['description'].fillna("")

items['Fav Recommendation Based on these']=items['director']+" "+items['description']

# 2. Convert text to TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(items['Fav Recommendation Based on these'])

# 3. Choose an item index to get recommendations for
item_index = 0   # "Red running shoes for men"

# 4. Compute cosine similarity between this item and all items
cos_sim = cosine_similarity(tfidf_matrix[item_index], tfidf_matrix).flatten()

# 5. Get top N similar items (excluding the item itself)
N = 5
similar_indices = np.argsort(-cos_sim)  # sort descending
similar_indices = [i for i in similar_indices if i != item_index][:N]

print("Base item:", items.loc[item_index,['director','description' ]])
print("\nTop recommendations:")
for i in similar_indices:
    print(f"- {items.loc[i, ['director', 'description']]} (score={cos_sim[i]:.3f})")
```
