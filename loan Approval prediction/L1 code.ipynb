```python
# chapter 10- project 1 from ML - loan approval prediction

import pandas as pd
data=pd.read_csv('loan_approval_dataset.csv')

data.columns=data.columns.str.strip()

#data cleaning
print(data.info())
data.dropna(inplace=True)
print(data.isnull().sum())
data.drop_duplicates(inplace=True)
print(data.duplicated().sum())

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=['loan_status','education','self_employed']
for col in cols:
    data[col]=le.fit_transform(data[col])

#data['loan_status']=le.fit_transform(data['loan_status'])

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
cols = ['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'residential_assets_value', 'commercial_assets_value', 'bank_asset_value']

# THIS SINGLE LINE fixes everything:
data[cols] = sc.fit_transform(data[cols])  # NO LOOP! 2D input ✓

X=data.drop('loan_status',axis=1)
y=data['loan_status']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
# 1.Naive bayes
from sklearn.naive_bayes import BernoulliNB
BNB=BernoulliNB()
BNB.fit(X_train,y_train)
y_pred_BNB=BNB.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy=accuracy_score(y_test,y_pred_BNB)
print(f'Accuracy: {accuracy * 100: .2f}%')

precision=precision_score(y_test,y_pred_BNB)
print(f'Precision: {precision * 100: .2f}%')

recall=recall_score(y_test,y_pred_BNB)
print(f'Recall: {recall * 100: .2f}%')

f1_score=f1_score(y_test,y_pred_BNB)
print(f'F1 Score: {f1_score * 100: .2f}%')

roc_auc=roc_auc_score(y_test,y_pred_BNB)
print(f'ROC AUC: {roc_auc * 100: .2f}%')
```

```python
# 2.Logistics Regression
from sklearn.linear_model import LogisticRegression
LR= LogisticRegression()
LR.fit(X_train,y_train)
y_pred_LR=LR.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy=accuracy_score(y_test,y_pred_LR)
print(f'Accuracy: {accuracy * 100: .2f}%')

precision=precision_score(y_test,y_pred_LR)
print(f'Precision: {precision * 100: .2f}%')

recall=recall_score(y_test,y_pred_LR)
print(f'Recall: {recall * 100: .2f}%')

f1_score=f1_score(y_test,y_pred_LR)
print(f'F1 Score: {f1_score * 100: .2f}%')

roc_auc=roc_auc_score(y_test,y_pred_LR)
print(f'ROC AUC: {roc_auc * 100: .2f}%')
```

```python
# 3.Support vector Machine
from sklearn.svm import SVC
SVM=SVC()
SVM.fit(X_train,y_train)
y_pred_SVM=SVM.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy=accuracy_score(y_test,y_pred_SVM)
print(f'Accuracy: {accuracy * 100: .2f}%')

precision=precision_score(y_test,y_pred_SVM)
print(f'Precision: {precision * 100: .2f}%')

recall=recall_score(y_test,y_pred_SVM)
print(f'Recall: {recall * 100: .2f}%')

f1_score=f1_score(y_test,y_pred_SVM)
print(f'F1 Score: {f1_score * 100: .2f}%')

roc_auc=roc_auc_score(y_test,y_pred_SVM)
print(f'ROC AUC: {roc_auc * 100: .2f}%')
```

```python
# 4.Decision Tree
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(X_train,y_train)
y_pred_DT=DT.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy=accuracy_score(y_test,y_pred_DT)
print(f'Accuracy: {accuracy * 100: .2f}%')

precision=precision_score(y_test,y_pred_DT)
print(f'Precision: {precision * 100: .2f}%')

recall=recall_score(y_test,y_pred_DT)
print(f'Recall: {recall * 100: .2f}%')

f1_score=f1_score(y_test,y_pred_DT)
print(f'F1 Score: {f1_score * 100: .2f}%')

roc_auc=roc_auc_score(y_test,y_pred_DT)
print(f'ROC AUC: {roc_auc * 100: .2f}%')
```

```python
# 5. Random Forest
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(X_train,y_train)
y_pred_RF=RF.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy=accuracy_score(y_test,y_pred_RF)
print(f'Accuracy: {accuracy * 100: .2f}%')

precision=precision_score(y_test,y_pred_RF)
print(f'Precision: {precision * 100: .2f}%')

recall=recall_score(y_test,y_pred_RF)
print(f'Recall: {recall * 100: .2f}%')

f1_score=f1_score(y_test,y_pred_RF)
print(f'F1 Score: {f1_score * 100: .2f}%')

roc_auc=roc_auc_score(y_test,y_pred_RF)
print(f'ROC AUC: {roc_auc * 100: .2f}%')
```
