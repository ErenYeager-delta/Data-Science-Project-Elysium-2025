```python
import os

"path = r'/content/drive/MyDrive/images/train'"
path = r'C:\Users\1771v\Downloads\dog-cat-full-dataset-master\data'
classes = os.listdir(path)
print(classes)

from tensorflow.keras.preprocessing import image_dataset_from_directory

base_dir = r'C:\Users\1771v\Downloads\dog-cat-full-dataset-master\data'
train_datagen = image_dataset_from_directory(base_dir, image_size=(200, 200),
                                             subset='training',
                                             seed=1,
                                             validation_split=0.1,
                                             batch_size=32)

test_datagen = image_dataset_from_directory(base_dir, image_size=(200, 200),
                                            subset='validation',
                                            seed=1,
                                            validation_split=0.1,
                                            batch_size=32)

import tensorflow as tf
from tensorflow import keras
from keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras.layers import Conv2D, MaxPooling2D

model = tf.keras.models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)),
    #input layer, #convolution layer (32)
    #relu == regression linear unit
    #input shape == 200,200,3(rgb)

    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),  #hidden layer

    layers.Dense(1, activation='sigmoid')  #output sigmoid === binary  (dog,cat) #softmax (multi)
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#categorical_crossentropy

# Train the model

model.fit(train_datagen, validation_data=test_datagen, epochs=10, batch_size=64)
from keras.preprocessing import image
import numpy as np

#Input image
test_image = image.load_img(r'/content/drive/MyDrive/images/test/dog/dog4.jpg', target_size=(200, 200))
#For show image
from matplotlib import pyplot as plt

plt.imshow(test_image)
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
result = model.predict(test_image)
print(result)
#Mapping result array with the main name list
if (result < 1):
    print("Dog")
else:
    print("Cat")
```
